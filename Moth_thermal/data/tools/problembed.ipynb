```python
import os
import glob
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib as plt
from skimage import io, color
# import cv2
# from PIL import Image, ImageOps
# from skimage.transform import resize

dir_problemed = Path('../Problemed_data')
list(dir_problemed.iterdir())
```

```python
df_problemed = pd.read_csv('../problemed.csv')
df_problemed.columns = ['source','fname', 'note']
df_problemed
```

```python
dir_NGwimgs = dir_problemed.joinpath('NGwimgs')
names_NGwings =  set(dir_NGwimgs.glob('*'))
names_NGwings = set(path.stem.split('_cropped')[0] for path in names_NGwings)

names_NGwings
```

```python
df_NGwings = pd.DataFrame(names_NGwings, columns=['fname'])
df_NGwings['source'] = df_NGwings.fname.str[:4]
df_NGwings['note'] = '側面'
df_NGwings
```

```python
df_problemed = pd.concat([df_problemed, df_NGwings]
                         , sort=False).reset_index(drop=True)
df_problemed
```

```python
df_problemed.to_csv(dir_problemed.joinpath('problemed.csv'))
```
